```python
from langchain.agents import create_agent
from langchain.agents.middleware import ModelFallbackMiddleware, SummarizationMiddleware
from langchain.tools import tool
from langchain.messages import SystemMessage,HumanMessage,AIMessage
```

```python
@tool
def scrape(jobrole: str,exp:int) -> str:
  """Scrapes a particular website to find matches for the given jobrole and experience."""
  return f"Found 2 matches for {jobrole} with and experience of {exp} years: 1) Production Analyst <link-here> 2) Software Developer Engineer <link-here>"


@tool
def send_mail(mail:str) -> str:
  """Sends a mail to the given email id regarding the job matches."""
  return f"Sent details to {mail}"


agent = create_agent(
  model = "groq:qwen/qwen3-32b",
  tools = [scrape,send_mail],
  system_prompt= "You are an expert career advisor. According to the specifications given by the user, find him a job role!"
)


response = agent.invoke({"messages": "Please find me job matches. My job role is Data Analyst and experience of 5 years.My email-id is vijay@gmail.com"})
response['messages'][-1].content
```
